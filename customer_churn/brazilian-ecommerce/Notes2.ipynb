```python
import pandas as pd
```

```python
order_items = pd.read_csv('olist_order_items_dataset.csv')
order_items.head()
```

```python
orders = pd.read_csv('olist_orders_dataset.csv')
orders.head()
```

```python
order_total = pd.merge(orders,order_items, on = 'order_id')
order_total.head()
```

```python
# sell difference between 2017 and 2018
order_total['date'] = pd.to_datetime(order_total['order_purchase_timestamp'])
```

```python
import datetime as dt
order2017 = order_total[order_total['date'].dt.year == 2017]
order2017['order_total'] = order2017['order_item_id']*order2017['price']


order2018 = order_total[order_total['date'].dt.year == 2018]
order2018['order_total'] = order2018['order_item_id']*order2018['price']
order2017.order_total.sum() - order2018.order_total.sum()
```

```python
order2017.shape
```

```python
customer1 = order2017.groupby('customer_id').sum()
customer1.head()
```

```python
customer2 = order2018.groupby('customer_id').sum()
customer2.index.nunique()+ customer1.index.nunique()
```

```python
d = pd.merge(customer1,customer2, on = 'customer_id', how = 'left')
d.head()
```
