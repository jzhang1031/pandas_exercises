```python
feature = pd.concat([number_of_transactions, red_or_not],axis = 1)
feature.head()
```

```python
# create response feature
import numpy as np
order2018 = orders[orders['date'] >= '2018-01-01 00:00:00']
order2018['target'] = np.ones(order2018.shape[0],dtype=int)
target = order2018.groupby('customer_id').sum().drop('status',axis = 1)
target.head()
```

```python
features = pd.merge(feature, target, how = 'left',on = 'customer_id')
features.head()
```

```python
# change NaN to 0
features = features.fillna(0)
features.index.is_unique
features.head()
```

```python
#save output data as a csv file

features.to_csv('features_data.csv')
```

```python
data = pd.read_csv('features_data.csv',index_col = 'customer_id')
data.head()
```

```python
X = data.drop('target', axis = 1)
y = data['target']

# train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
predictions = logmodel.predict(X_test)
```

```python
#evaluation
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
